# Arrays (Lists) - Basics

Structure: Dynamic arrays.

Stack Operations:

append() to push.

pop() to remove from end.

Indexing:

Read/Write is O(1).

Negative Indexing: -1 accesses the last value.

Slicing: arr[1:3] gets values from index 1 up to (but not including) 3.

Initialization: arr = [1] * 5 creates [1, 1, 1, 1, 1].

```python
# Arrays (Lists) - Basics

# 1. Initialization (Dynamic array)
arr = [1] * 5
print("Initial array:", arr)  
# Output: [1, 1, 1, 1, 1]

# 2. Append (Push to stack)
arr.append(6)
print("After append:", arr)  
# Output: [1, 1, 1, 1, 1, 6]

# 3. Pop (Remove from end - stack behavior)
removed_value = arr.pop()
print("Removed value:", removed_value)  
print("After pop:", arr)  
# Output: 6
#         [1, 1, 1, 1, 1]

# 4. Indexing (O(1) read/write)
print("Value at index 0:", arr[0])  # Read

arr[0] = 10  # Write
print("After modifying index 0:", arr)

# 5. Negative Indexing
print("Last value using -1:", arr[-1])

# 6. Slicing
slice_example = arr[1:3]
print("Slice arr[1:3]:", slice_example)
# Gets elements at index 1 and 2
```

## Arrays (Lists) - Advanced
Looping:

Value only: for n in nums:

Index & Value: for i, n in enumerate(nums):

Multiple Arrays: for n1, n2 in zip(nums1, nums2):

Sorting:

arr.sort() sorts ascending by default.

arr.sort(reverse=True) for descending.

Custom Sort: Use lambda functions (e.g., key=lambda x: len(x)).

List Comprehension: Concise syntax to create lists (e.g., [i + i for i in range(5)]).

2D Arrays Warning: Do not use [[0]*4]*4 (creates shallow copies). Use list comprehension: [[0]*4 for i in range(4)].

```python
# Initial Data
students = ["Alice", "Bob", "Charlie", "David"]
scores = [88, 92, 78, 95]

### 1. Looping: zip() and enumerate()
print("--- Student Rankings ---")
# Using zip() to iterate over two lists simultaneously
# Using enumerate() to track the rank (index)
for i, (name, score) in enumerate(zip(students, scores), start=1):
    print(f"Rank {i}: {name} scored {score}")


### 2. List Comprehension
# Let's give everyone a 5-point curve, but only if they haven't maxed out at 100
curved_scores = [min(100, s + 5) for s in scores]
print(f"\nCurved Scores: {curved_scores}")


### 3. Sorting (Standard and Custom)
# Sort names alphabetically (ascending)
students.sort()
print(f"Alphabetical: {students}")

# Custom Sort: Sort students by the length of their name using a lambda
students.sort(key=lambda name: len(name))
print(f"Sorted by name length: {students}")

# Reverse Sort: Sort scores descending
curved_scores.sort(reverse=True)
print(f"Top Scores: {curved_scores}")


### 4. 2D Arrays (The Right Way)
# We need a 3x3 grid to represent a classroom seating chart
# WRONG: seating_chart = [[None]*3]*3  <-- Editing one row would edit them all!

# RIGHT: Use list comprehension for independent rows
seating_chart = [["Empty" for _ in range(3)] for _ in range(3)]

# Now we can safely update one seat without affecting others
seating_chart[0][0] = "Alice"

print("\nSeating Chart Row 1:", seating_chart[0])
print("Seating Chart Row 2:", seating_chart[1])
```

## Strings
Immutability: Strings cannot be modified (e.g., s[0] = 'a' fails). Modifying creates a new string (O(n) operation).

Conversions: convert between ints and strings easily (int("123"), str(123)).

ASCII: Use ord("a") to get the ASCII value.

Join: Use "".join(list_of_strings) to combine strings efficiently.

```python
# 1. Immutability: The "Gotcha"
s = "Python"
print(f"Original: {s}")

try:
    s[0] = "p"  # This will trigger a TypeError
except TypeError as e:
    print(f"❌ Error: {e} (You can't change a string in place!)")

# To "modify" it, we must create a brand new string
s = "p" + s[1:] 
print(f"New string created: {s}")


# 2. Conversions & ASCII
# Let's turn a character into its numeric code and back
char = "A"
ascii_val = ord(char)  # Get ASCII (65)
next_char = chr(ascii_val + 1)  # Get "B"

print(f"\nASCII of '{char}' is {ascii_val}. The next letter is '{next_char}'.")


# 3. Efficient Joining
# Problem: We have a list of numbers as strings and want to merge them
parts = ["100", "200", "300", "400"]

# AVOID THIS: result = "" | for p in parts: result += p 
# (That creates a new string every loop! Very slow at scale)

# DO THIS: Use .join() (O(n) efficiency)
combined = "-".join(parts)
print(f"Joined string: {combined}")


# 4. Conversion Logic
# Let's extract the first number, turn it into an int, and double it
first_num = int(parts[0]) 
print(f"Math on converted string: {first_num} * 2 = {first_num * 2}")
```

## Hash Maps & Sets
Hash Sets:

No duplicates.

Search/Insert/Remove are O(1).

Initialization: myset = set() or Set comprehension.

Hash Maps (Dictionaries):

Key-Value pairs (no duplicate keys).

Operations: map["key"] = val, pop("key").

Looping: Loop through keys, .values(), or .items() (for key+value).

Dict Comprehension: { i : i*2 for i in range(3) }.

```python
### 1. Hash Sets (No Duplicates)
# Let's track unique IP addresses
visitor_ips = {"192.168.1.1", "10.0.0.5", "192.168.1.1"} # Notice the duplicate?

# Sets automatically prune duplicates
print(f"Unique IPs: {visitor_ips}")  # {'192.168.1.1', '10.0.0.5'}

# Set Comprehension: Create a set of square numbers (only unique values)
unique_squares = {x*x for x in [-2, -1, 0, 1, 2]}
print(f"Unique Squares: {unique_squares}") # {0, 1, 4}


### 2. Hash Maps (Dictionaries)
# Store usernames and their login counts
user_logins = {"Alice": 5, "Bob": 12}

# Update or Add: O(1) time
user_logins["Charlie"] = 1  # Add new
user_logins["Alice"] += 1   # Update existing

# Remove a key
user_logins.pop("Bob")


### 3. Looping through Dictionaries
print("\n--- User Dashboard ---")
# .items() is the gold standard for looping through both Key and Value
for user, count in user_logins.items():
    print(f"User: {user} | Logins: {count}")

# .values() if you only care about the data, not the keys
total_logins = sum(user_logins.values())
print(f"Total Logins across site: {total_logins}")


### 4. Dictionary Comprehension
# Let's create a map of numbers to their cubes
cubes = {i: i**3 for i in range(5)}
print(f"\nCube Map: {cubes}")
```

## Advanced Data Structures
Tuples:

Like arrays but immutable (cannot change values).

Key Usage: Can be used as keys in Hash Maps/Sets (Lists cannot).

Queues:

Use collections.deque.

Double-ended: append(), popleft() (O(1)), appendleft(), pop().

Heaps:

Use heapq. Implemented as Min-Heaps by default.

Max-Heap Workaround: Multiply values by -1 when pushing and popping.

heapify(arr): Converts a list to a heap in linear time O(n).

```python
from collections import deque
import heapq

### 1. Tuples: The Immutable Key
# Tuples are great for fixed data, like (x, y) coordinates.
# Unlike lists, they can be keys in a dictionary!
locations = {}
point_a = (10, 20)
locations[point_a] = "Main Entrance"

print(f"Location at {point_a}: {locations[point_a]}")


### 2. Queues: Double-Ended efficiency (deque)
# Use a deque when you need to add to one side and remove from the other.
task_queue = deque(["Email Client", "Update Server"])

task_queue.append("Backup Data")      # Add to end
first_task = task_queue.popleft()     # Remove from front O(1)

print(f"\nFinished: {first_task}")
print(f"Remaining in Queue: {list(task_queue)}")


### 3. Heaps: The Priority Machine (Min-Heap)
# By default, heapq always gives you the SMALLEST number first.
grades = [88, 95, 70, 100, 82]
heapq.heapify(grades)  # Turns list into heap in O(n)

print(f"\nLowest Grade (Min-Heap): {heapq.heappop(grades)}") # 70


### 4. Max-Heap Workaround
# To get the BIGGEST number first, multiply by -1.
max_heap = []
scores = [10, 50, 20]

for s in scores:
    heapq.heappush(max_heap, -s) # Push negative values

# When popping, multiply by -1 again to get the original number
top_score = -heapq.heappop(max_heap)
print(f"Highest Score (Max-Heap): {top_score}") # 50
```

Pro-Tip: Never use list.pop(0) for a queue. It forces Python to shift every other element in the list one spot to the left, which is $O(n)$. Always use collections.deque for $O(1)$ performance.

## Section 6: File Input and Output
Read from and write to files using open(), read(), write(), and context managers with the 'with' statement.

```python
# Writing to a file
filename = "sample.txt"

# Using 'with' statement (context manager)
with open(filename, 'w') as file:
    file.write("Hello, Python!\n")
    file.write("This is a sample file.\n")
    file.write("File I/O is easy!\n")

print(f"File '{filename}' created successfully.")

# Reading from a file
print(f"\nReading from file '{filename}':")
with open(filename, 'r') as file:
    content = file.read()
    print(content)

# Reading line by line
print("Reading line by line:")
with open(filename, 'r') as file:
    for line in file:
        print(f"  {line.strip()}")

# Appending to a file
with open(filename, 'a') as file:
    file.write("This line was appended.\n")

print(f"\nLine appended to '{filename}'.")
```

## Section 7: Error Handling and Exceptions
Use try/except/finally blocks to handle exceptions and understand common error types in Python.

```python
# Error Handling - ValueError
print("Example 1: ValueError")
try:
    num = int("not_a_number")
except ValueError:
    print("  Error: Could not convert string to integer")

# Error Handling - ZeroDivisionError
print("\nExample 2: ZeroDivisionError")
try:
    result = 10 / 0
except ZeroDivisionError:
    print("  Error: Cannot divide by zero")

# Error Handling - IndexError
print("\nExample 3: IndexError")
try:
    my_list = [1, 2, 3]
    value = my_list[10]
except IndexError:
    print("  Error: Index out of range")

# Error Handling - Multiple exceptions
print("\nExample 4: Multiple exceptions")
def safe_divide(a, b):
    try:
        result = int(a) / int(b)
        return result
    except ValueError:
        print("  Error: Invalid input for conversion to integer")
    except ZeroDivisionError:
        print("  Error: Cannot divide by zero")
    finally:
        print("  Safe divide function completed")

safe_divide(10, 2)
safe_divide(10, 0)
safe_divide("ten", 2)
```

## Section 8: Object-Oriented Programming Basics
Define classes, create objects, use constructors and methods, and understand inheritance and encapsulation principles.

```python
# Define a class
class Dog:
    """A simple Dog class"""
    
    # Class variable
    species = "Canis familiaris"
    
    # Constructor
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    # Methods
    def bark(self):
        return f"{self.name} says: Woof!"
    
    def get_info(self):
        return f"{self.name} is {self.age} years old"

# Create objects
dog1 = Dog("Buddy", 3)
dog2 = Dog("Max", 5)

print(f"Dog 1: {dog1.get_info()}")
print(f"Dog 2: {dog2.get_info()}")
print(f"\n{dog1.bark()}")
print(f"{dog2.bark()}")
print(f"\nSpecies: {Dog.species}")
```

```python
# Inheritance
class Animal:
    """Parent class"""
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return f"{self.name} makes a sound"

class Cat(Animal):
    """Child class inheriting from Animal"""
    def speak(self):
        return f"{self.name} says: Meow!"

class Bird(Animal):
    """Another child class"""
    def speak(self):
        return f"{self.name} says: Tweet!"

# Create objects from child classes
cat = Cat("Whiskers")
bird = Bird("Tweety")

print("Inheritance Example:")
print(cat.speak())
print(bird.speak())

# Encapsulation
print("\n\nEncapsulation Example:")
class BankAccount:
    """Example of encapsulation"""
    def __init__(self, balance=0):
        self.__balance = balance  # Private attribute
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"Deposited ${amount}. New balance: ${self.__balance}"
        return "Invalid amount"
    
    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            return f"Withdrew ${amount}. New balance: ${self.__balance}"
        return "Invalid amount or insufficient funds"
    
    def get_balance(self):
        return f"Current balance: ${self.__balance}"

account = BankAccount(1000)
print(account.get_balance())
print(account.deposit(500))
print(account.withdraw(200))
print(account.get_balance())
```
